### Step 3 - Explore the mRNA data. 

The first of the three questions we want to answer above is "How variable is gene expression data?". This is an important question to answer when you start looking at this type of high-dimensional data. When you compare two biological states, you want to be able to know at what threshold you can call expression "different" between those two states - and unless you know how variable gene expression data can be, you won't be able to do that. If someone tells us that gene 1 in condition A is 1.5 times the level of gene 1 in condition B, how do we know if this is really different, or simply what we might expect from normal biological variation? 

Let us first look at the correlation between these mRNA replicates. Run the code below, without modification. 


```r
# use ggplot to make a scatter plot of mRNA1 vs mRNA2 from the object expression_data. 
# set the x and y axes to log scale, and add a linear regression line 
ggplot(expression_data, aes(mRNA1, mRNA2)) + geom_point() + scale_y_log10() + scale_x_log10() +stat_smooth(method = "lm")

# perform linear regression on the mRNA data and calculate the Pearson R squared value
m <- lm(mRNA2 ~ mRNA1, expression_data);
r2 = format(summary(m)$r.squared, digits = 3)
cat("R squared is",r2)
```

Correlation is a good way of seeing how similar data sets are to each other. But we are interested in differences, so this way of displaying the data is not very helpful. Let's transform the data into a measure that we often see in gene expression comparisons - the log fold change. 

Log fold change (or logFC for short) is simply the Log2 of the ratio of one gene measure over another gene measure. eg. log2(mRNA1/mRNA2). The ratio of mRNA1/mRNA2 provides us with the measure of how much something has changed. If mRNA1 is 3 and mRNA2 is 2, then the fold change (mRNA1/mRNA) is 1.5 times. 

By placing those ratios on a log2 scale, we are making it so that numerically, a ratio of less than 1 is the same as a ratio of greater than 1. That is because log2(2/1) is 1, and log2(1/2) is -1. This makes intuitive sense for biologists, as a doubling is equally important biologically as a halving, but without the log transformation, we would have 2 vs 0.5, which are not equal distances away from the centre (1/1). 

Let's calculate the logFC for the mRNA data, and then plot the results in a histogram, so that we can see what magnitude of change is usual for mRNA data. 



```r
#create a new column called mRNA_logFC in the object called expression data, and add to that the log base 2 of mRNA1/mRNA2
expression_data$mRNA_logFC <- log(expression_data$mRNA1/expression_data$mRNA2,2)

#create a histogram of the data in column mRNA_logFC of the object expression data, dividing the data into 20 bins. 
ggplot(expression_data, aes(mRNA_logFC)) + geom_histogram(bins=20)
```
